```sql
CREATE WIDGET TEXT catalogo default "";
```

```sql
CREATE CATALOG IF NOT EXISTS $catalogo;
```

```sql
CREATE SCHEMA IF NOT EXISTS $catalogo.bronze;
CREATE SCHEMA IF NOT EXISTS $catalogo.silver;
CREATE SCHEMA IF NOT EXISTS $catalogo.golden;
```

###Tablas Bronze

```sql
CREATE TABLE $catalogo.bronze.orders(
  order_id INT,
  user_id INT,
  eval_set STRING,
  order_number INT,
  order_dow INT,
  order_hour_of_day INT,
  days_since_prior_order DOUBLE,
  ingestion_date TIMESTAMP
) 
USING DELTA;
```

```sql
CREATE TABLE $catalogo.bronze.order_products(
  order_id INT,
  product_id INT,
  add_to_cart_order INT,
  reordered INT,
  ingestion_date TIMESTAMP
) 
USING DELTA;
```

```sql
CREATE TABLE $catalogo.bronze.products(
  product_id INT,
  product_name STRING,
  aisle_id INT,
  department_id INT,
  ingestion_date TIMESTAMP
) 
USING DELTA;
```

```sql
CREATE TABLE $catalogo.bronze.aisles(
  aisle_id INT,
  aisle STRING,
  ingestion_date TIMESTAMP
) 
USING DELTA
```

```sql
CREATE TABLE $catalogo.bronze.departments(
  department_id INT,
  department STRING,
  ingestion_date TIMESTAMP
) 
USING DELTA;
```

###Tablas Silver

```sql
CREATE TABLE $catalogo.silver.ventas_detalle(
  order_id INT,
  user_id INT,
  eval_set STRING,
  order_number INT,
  order_dow INT,
  order_hour_of_day INT,
  days_since_prior_order DOUBLE,
  product_id INT,
  product_name STRING,
  add_to_cart_order INT,
  reordered INT,
  aisle_id INT,
  aisle STRING,
  department_id INT,
  department STRING,
  franja_horaria STRING,
  dia_semana STRING,
  es_recompra BOOLEAN,
  ingestion_date TIMESTAMP
)
USING DELTA
```

###Tablas Golden

```sql
CREATE TABLE $catalogo.golden.ventas_por_departamento(
  department STRING,
  total_productos_pedidos LONG,
  total_ordenes LONG,
  total_recompras LONG,
  pct_recompra DOUBLE,
  promedio_productos_por_orden DOUBLE,
  ingestion_date TIMESTAMP
) 
USING DELTA;
```

```sql
CREATE TABLE $catalogo.golden.top_productos(
  product_id INT,
  product_name STRING,
  aisle STRING,
  department STRING,
  total_pedidos LONG,
  total_recompras LONG,
  pct_recompra DOUBLE,
  rank_popularidad INT,
  ingestion_date TIMESTAMP
) 
USING DELTA;
```

```sql
CREATE TABLE $catalogo.golden.frecuencia_usuarios (
  user_id INT,
  total_ordenes LONG,
  promedio_dias_entre_ordenes DOUBLE,
  total_productos_distintos LONG,
  tipo_cliente STRING,
  ingestion_date TIMESTAMP
) 
USING DELTA;
```
